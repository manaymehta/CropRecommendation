```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DS3.csv")
df = df.drop(["District_Name", "Crop_Year", "Area"], axis = "columns")
```

```python
from sklearn.preprocessing import  StandardScaler
scaler = StandardScaler()
scaler = StandardScaler()
df['Production'] = scaler.fit_transform(df[['Production']])
df['Best_Crop'] = df.groupby(['State_Name', 'Season'])['Production'].transform(lambda x: df.loc[x.idxmax(), 'Crop'])
df = df.drop_duplicates(subset=['State_Name', 'Season'], keep='first')
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

label_encoders = {}
for col in ['State_Name', 'Season', 'Best_Crop']:
    le = LabelEncoder()
    df[f'{col}_n'] = le.fit_transform(df[col])
    label_encoders[col] = le

X = df[['State_Name_n', 'Season_n']]
y = df[['Best_Crop_n']]

print(X)
print(y)
print(label_encoders)
```

```python
state_map = dict(zip(df['State_Name'], df['State_Name_n']))
season_map = dict(zip(df['Season'], df['Season_n']))
crop_map = dict(zip(df['Best_Crop_n'], df['Best_Crop']))

# Print the mappings
print("State Name Encoding:", state_map)
print("Season Encoding:", season_map)
print("Best Crop Encoding:", crop_map)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python
state_input = input("Enter State Name: ")
season_input = input("Enter Season: ")
state_encoded = state_map.get(state_input)
season_encoded = season_map.get(season_input)

input_data = [[state_encoded, season_encoded]]
predicted_crop_encoded = clf.predict(input_data)[0]
best_crop_output = crop_map.get(predicted_crop_encoded)
print(best_crop_output)
```
